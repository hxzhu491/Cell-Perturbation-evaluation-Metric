## Output input.h5ad to csv in each folder

```python
import scanpy as sc
import numpy as np
import pandas as pd
```

```python
input_B = sc.read_h5ad("cell_type_B/input.h5ad")
input_Myeloid = sc.read_h5ad("cell_type_Myeloid/input.h5ad")
input_NK = sc.read_h5ad("cell_type_NK/input.h5ad")
input_T4 = sc.read_h5ad("cell_type_T4/input.h5ad")
input_T8 = sc.read_h5ad("cell_type_T8/input.h5ad")
input_Treg = sc.read_h5ad("cell_type_Treg/input.h5ad")
```

```python
input_B.obs.to_csv("input_obs/input_B.csv")
input_Myeloid.obs.to_csv("input_obs/input_Myeloid.csv")
input_NK.obs.to_csv("input_obs/input_NK.csv")
input_T4.obs.to_csv("input_obs/input_T4.csv")
input_T8.obs.to_csv("input_obs/input_T8.csv")
input_Treg.obs.to_csv("input_obs/input_Treg.csv")
```
